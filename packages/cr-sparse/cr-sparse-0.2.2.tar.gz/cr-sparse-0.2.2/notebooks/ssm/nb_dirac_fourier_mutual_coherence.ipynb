# Mutual Coherence between Standard Basis and Fourier Basis

```python
import jax
import jax.numpy as jnp
import numpy as  np
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import cr.sparse.dict as crdict
```

```python
N  = 128
```

```python
def dirac_fourier_mutual_coherence(N):
    I = jnp.eye(N)
    F = crdict.fourier_basis(N)
    return crdict.mutual_coherence(I, F)
    
```

```python
dirac_fourier_mutual_coherence(4)
```

```python
Ns = jnp.arange(2, 513,2)
```

```python
mus = [dirac_fourier_mutual_coherence(n) for n in Ns]
```

```python
fig_params = dict(figsize=(8,8), dpi= 100, facecolor='w', edgecolor='k')
fig, ax =plt.subplots(**fig_params)
ax.plot(Ns, mus);
ax.set_title('Mutual coherence between Dirac basis and Fourier basis for $\mathbb{C}^N$');
ax.set_xlabel('N (Complex vector space dimension) ');
ax.set_ylabel('$\mu$ (Mutual coherence)');
ax.grid();
```
